```python
#인구 밀도가 10000이 넘는 도시는 총 몇개?( 인구밀도 = 인구수/땅의 면적) 
condition = world_df['Population'] / world_df['Land area (in sqKm)'] > 10000
world_df['10000이 넘는 도시 수']= condition
world_df.iloc[:,4].value_counts()
```

```python
## 정답
world_df["Density"] = world_df['Population'] / world_df['Land area (in sqKm)']
df_high_density = world_df[world_df["Density"]> 10000]
df_high_density
```

```python
# 인구 밀도가 가장 높은 도시
world_df.sort_values("Density")
```

```python
##정답
density_ranks = world_df.sort_values(by = "Density", ascending = False)
density_ranks.iloc[:,0]
```
